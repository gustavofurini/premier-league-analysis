## **Merging all the tables with teams stats**


```python
dfs = ['touches_in_opp_box_team', 'corner_taken_team', 'fk_foul_lost_team', 'accurate_long_balls_team', 'won_tackle_team', 'accurate_cross_team',
        'clean_sheet_team', 'goals_conceded_team_match', 'total_yel_card_team', 'possession_percentage_team', 'expected_goals_conceded_team',
        'interception_team', 'big_chance_missed_team', 'total_red_card_team', 'team_ratings', 'penalty_conceded_team', 'penalty_won_team',
        'expected_goals_team','team_goals_per_match', 'accurate_pass_team', 'saves_team', 'possession_won_att_3rd_team', 'big_chance_team', 
        'effective_clearance_team', 'ontarget_scoring_att_team']
tables = {}

# Loop para carregar cada tabela
for table_names in dfs:
    table_path = f'{path}/{table_names}.csv'  # Altere a extensão se necessário
    tables[table_names] = pd.read_csv(table_path) 
```

```python
for table_names, df in tables.items():
    print(f"{table_names}: {df.shape}")
```

```python
import pandas as pd

merged_df = list(tables.values())[0]

for nome_tabela, df in list(tables.items())[1:]:
    merged_df = pd.merge(
        merged_df, df, on='Team', how='outer', suffixes=('', f'_{nome_tabela}')
    )

for col in merged_df.columns:
    if col.endswith(tuple(tables.keys())):  
        merged_df = merged_df.drop(columns=[col])

merged_df.to_csv('table_teams.csv', index=False)
```
